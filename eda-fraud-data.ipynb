```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
fraud_data = pd.read_csv('data/raw/Fraud_Data.csv')
print(fraud_data.info())
print(fraud_data.describe())
print(fraud_data['class'].value_counts())
```

```python
# Visualize class imbalance
sns.countplot(x='class', data=fraud_data)
plt.title("Fraud vs Non-Fraud Transactions")
plt.show()
```

```python
# Numeric feature distributions
numeric_features = ['purchase_value', 'age']
fraud_data[numeric_features].hist(bins=30, figsize=(10,5))
plt.show()
```

```python
# Fraud vs purchase_value
sns.boxplot(x='class', y='purchase_value', data=fraud_data)
plt.show()
```

```python

# Fraud vs categorical features
sns.countplot(x='browser', hue='class', data=fraud_data)
plt.title("Fraud by Browser Type")
plt.show()
```
